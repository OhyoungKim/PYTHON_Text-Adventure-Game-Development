```python
# GAME_ANOTHER_STORY_ALADDIN designed by Ohyoung Kim

"""
    Docstring:

    A) Introduction:
    This game is based on Disney film, Aladdin (1992). It consists of 3 stages
    and has defined functions for start, win, and lose. It requires wisdom 
    and a bit of good luck. As a hidden helper not mentioned in the fairy 
    tale, help Aladdin to follow his fate in the story.
    
    Stage 1: Pub - Get the clue to where the lamp is
    Stage 2: In front of a cave - Unlock the cave
    Stage 3: In front of the last chamber of the cave - Solve the riddle of the Sphinx
    
    B) Known Bugs and/or Errors:
    None.

"""

# Import things to make it function
import sys
import random
from time import sleep


# Establishing base for running game
def runGame():
    global protagonist_name
    global host_name
    global user_name

    # Introducing the game background story & purpose
    print(f"""
    {'-' * 23} ANOTHER STORY - ALADDIN {'-' * 23}\n
    The fairy tale stories that the world knows are twisted and reversed. 
    A fairy tale with a protagonist unable to live up to his given fate will 
    disappear. Please help Aladdin to follow his fate in the story!
    \n    {'-' * 72}\n""")
    
    # Declaring protagonist_name
    protagonist_name = "ALADDIN"
    
    # Start a game when user press ENTER (Timer starts)
    input(prompt = "\n<Press ENTER to start game>\n")
    
    # Declaring host_name (Anagram: MADRE == DREAM)
    host_name = "MADRE"
    
    print(f"""{host_name}: Hi, I'm the one who manages fairy tales, called {host_name}. 
       What is your name?""")
    
    # Establishing an input to obtain a username 
    user_name = input(prompt = "> ")
    # Capitalizing user_name
    user_name = user_name.upper()
        
    print(f"""\n{host_name}: Ok, {user_name}, please look after {protagonist_name}! Good Luck Have Fun!!\n\n""")    
    # Controlling the speed of user engagement
    input(prompt = "\n<Press ENTER to continue>\n")
    stage_1()  # Moving to stage 1
    

    
######################################################
## Stage 1: Pub - Get the clue to where the lamp is ##
######################################################

def stage_1():
    print('-' * 71)
    print(f"""\n{protagonist_name}: {user_name}, put it together!\n""")
    print(f"""{user_name}: ...What is it?""")
    
    sleep(0.7)  # Establishing a pause
    print("""   ......Aladdin?!\n""")    
    print(f"""{protagonist_name}: Hey, what's up with you?""")
    print(f"""         Anyway, listen, {user_name}. Rumor has it that a STRANGER with a clue to 
         where the lamp is opened a game. 
         Let's be the winner of the game and get the clue!""")
    
    # Controlling the speed of user engagement
    input(prompt = "\n\n<Press ENTER to talk to stranger>\n")
    
    print(f"""{protagonist_name}: Hey stranger, let's play with us!!\n""")
    
    print("""STRANGER: Accept your challenge! But it is a two-player game. 
          Decide whichever of you plays the game with me.\n\n""")
           
    # Looping until a user input correctly
    while True:
        # Giving the choice between Aladdin & User
        choice = input(prompt = f"""Who plays game with a stranger?\n
    1) Aladdin
    2) I (User)\n
    > """)
        # Converting lower case
        choice = choice.lower()
        
        # Conditional statement to move to Aladdin's side or user's side                 
        if "aladdin" in choice or "1" in choice:
            print(f"""\n{protagonist_name}: ...Sorry..I've got a tummy ache. Please play the game for me.""")
            print(f"""{user_name}: God.. Taking a break. Let me do it.\n""")
            # Establishing a pause
            input(prompt = "\n<Press ENTER to start stranger's game>\n")
            game_1()  # Moving to game_1
            
        elif "i" in choice or "user" in choice or "2" in choice:
            print(f"""\n{protagonist_name}: Great choice! I cheer as best as I can!!\n""")
            # Establishing a pause
            input(prompt = "\n<Press ENTER to start stranger's game>\n")
            game_1()  # Moving to game_1
            
        # Controlling unexpected inputs    
        else:
            print(f"""\n{host_name}: You know that is not an option... Let's try again.\n""")
            sleep(0.7) # Establishing a pause

            
def game_1():
    # Introducing the game: Baskin Robbins 31
    print("""
STRANGER: Let's start the game named Baskin Robbins 31. 
          We take turns and count freely from one to three numbers. 
          As a result, the one who says 31 loses. You first.""")
    
    # Establishing start for the game
    number = 0    
    turn = 0 
    
    # Looping until a user input correctly
    while True:
        # Establishing for user's turn
        if turn == 0:
            # Establishing an input to play game and converting it for numerical comparison
            user = int(input(prompt = "\nPlease input from 1 to 3: "))
            
            # Controlling unexpected inputs
            if user != 1 and user != 2 and user != 3:
                print(f"""\nSTRANGER: Are you kidding me? You can call from 1 to 3!""")
                continue
            
            # Counting user's number - Loop based on user input
            for num in range(user):
                number += 1
                print(f"{user_name}:", number)
                
                # Establishing final number
                if number == 31:
                    break
                
            # Making other's turn    
            turn += 1
            # Getting user's turn
            turn %= 2
        
        
        # Establishing for stranger's turn
        elif turn == 1:
            # Establishing stranger's call from 1 to 3
            s = random.randint(1, 3)
            
            # Counting stranger's number - Loop based on established input
            for num in range(s):
                number += 1
                print('STRANGER:', number)
                
                # Establishing final number
                if number == 31:
                    break
            
            # Making other's turn 
            turn += 1
            # Getting stranger's turn
            turn %= 2
        
        # Finishing the game
        if number >= 31:
            break
    
    sleep(1.0)  # Establishing a pause
    
    # Establishing that user wins
    if turn == 0: 
        print("""\nSTRANGER: Wow, how clever you are! Take the clue :) \n""")
        print(f"""{protagonist_name}: I knew you would win, {user_name}! Let's follow the clue!""")
        
        # Controlling the speed of user engagement
        input(prompt = "\n\n<Press ENTER to follow the clue>\n")
        stage_2()  # Moving to stage 2
        
    # Establishing that stranger wins
    else: 
        print("""\nSTRANGER: Nice try! 
          Even though you lost the game, I give you the clue :) \n""")
        print(f"""{protagonist_name}: Thanks!! Let's follow the clue, {user_name}!!!""")
        
        # Controlling the speed of user engagement
        input(prompt = "\n\n<Press ENTER to follow the clue>\n")
        stage_2()  # Moving to stage 2
    

    
###################################################
## Stage 2: In front of a cave - Unlock the cave ##
###################################################

def stage_2():
    print('-' * 71)
    print(f"""\n{protagonist_name}: Finally, we're here! The lamp is in this cave!\n""")
    print(f"""{user_name}: Oh, wait! There's something at the entrance!!\n""")
    print(f"""{protagonist_name}: At the entrance?\n""")

    # Controlling the speed of user engagement
    input(prompt = "\n<Press ENTER to check the entrance of the cave>\n\n")
    
    print(f"""{protagonist_name}: ......Lock?!!!!\n""")
    
    sleep(0.7) # Establishing a pause
    print("MAGIC LOCK: Please enter the password between 10 and 50 within 7 times.\n")
    print(f"""{protagonist_name}, {user_name}: WHAT?!!!!!\n""")
    
    # Establishing the password between 10 and 50
    real_pw = random.randint(10, 50)
        
    # Looping with limited attempts    
    for i in range(7):
        # Establishing an input to play game and converting it for numerical comparison
        guess_pw = int(input(prompt = "\nGuess the number for password (10 ~ 50): "))
        
        if real_pw == guess_pw:
            print("\n\nMAGIC LOCK: The lock has been released.\n")
            print(f"""\n{protagonist_name}: You're amazing!! Let's go inside the cave!!!""")
            
            # Controlling the speed of user engagement
            input(prompt = "\n\n<Press ENTER to go into the cave>\n")
            stage_3()  # Moving to stage 3
            
        # Setting different error messages for each case    
        elif real_pw > guess_pw:
            print("\nMAGIC LOCK: Invalid password. Password is larger than you guess.\n")
            
        else:
            print("\nMAGIC LOCK: Invalid password. Password is smaller than you guess.\n")
        
    # Establishing what happens when all the attemps are over
    if i == 6:
        print("""\n\nMAGIC LOCK: You've used all the attemps. Password for this time was {}""".format(real_pw))
        print(f"""\n\n \t{'-' * 30} GAME OVER {'-' * 30}\n""")
        fail()  # Game over
    

    
###############################################################################
## Stage 3: In front of the last chamber of the cave - Solve the riddle of the Sphinx ##
###############################################################################

def stage_3():
    print('-' * 71)
    print(f"""\n{protagonist_name}: The cave is deep. Where the hell is the lamp?\n""")
    
    sleep(1.3)  # Establishing a pause
    print("\n?????: New challenger?\n")
    print(f"""{protagonist_name}: A statue speaks?!\n""")
    print(f"""{user_name}: By the gods! It's not a statue, {protagonist_name}! It's the Sphinx!!!\n""")
    
    print("""SPHINX: Once again, are you a new challenger for my riddles?
        If you can answer my questions wisely, I will move for you to pass.\n""")
    
    # Looping until a user input correctly
    while True:
        # Giving the choice between hearing riddles & running away
        choice = input(prompt = """Would you like to accept the Sphinx's riddles?
        1) Yes, can I hear the riddles?
        2) N..no! I,I,I think I would better to find another way...\n
        > """)
        # Converting lower case
        choice = choice.lower()
        
        # Conditional statement to continue or fail
        if "yes" in choice or "hear" in choice or "1" in choice:
            print("""\nSPHINX: Good! No need to get scared in advance, challenger.\n""")

            # Establishing a pause
            input(prompt = "\n<Press ENTER to hear the riddles>\n")
            
            print("""SPHINX: Which creature has one voice and yet becomes four-footed and two-footed 
        and three-footed?\n""")            
            print(f"""{protagonist_name}: I know it!! It's MAN!!!!! 
         A MAN crawls on all fours as a baby, then walks on two feet as an adult, 
         and then uses a walking stick in old age.\n""")
            
            # Controlling the speed of user engagement
            input(prompt = "\n<Press ENTER to continue>\n")
            
            print("""SPHINX: Exactly. 
        There is a second riddle. There are two sisters: one gives birth to the 
        other and she, in turn, gives birth to the first. Who are the sisters?\n""")
            print(f"""{protagonist_name}, {user_name}: Umm...\n""")
                       
            sleep(3.0) # Setting timer for pause 
            
            print("""\nSPHINX: ...Any idea? 
        Umm..I don't want to end it like this... 
        Ok, I give you a hint. Pay attention to capitals.\n""")            
            
            # Establishing a pause
            input(prompt = "\n<Press ENTER to get a hint>\n")    
            game_3()  # Moving to game_3
            
            # Controlling the speed of user engagement            
            input(prompt = "\n\n<Press ENTER to solve the riddle>\n\n")
            
            print("""SPHINX: Can you answer now?\n""")
            print(f"""{protagonist_name}: Yes!! DAY and NIGHT!!!!!\n""")
            print("""SPHINX: Exactly. Go on and get the lamp :) \n""")
            
            # Controlling the speed of user engagement
            input(prompt = "\n<Press ENTER to continue>\n")
            
            print(f"""\n{host_name}: It's all you can involve in the Aladdin's story. 
       Thanks for helping Aladdin out. It's time to wake up now!""")
            print(f"""\n\n{'-' * 30} THE END {'-' * 30}""")
            sys.exit() # The end     
            
        
        elif "no" in choice or "another" in choice or "2" in choice:
            print(f"""\n \t{'-' * 30} GAME OVER {'-' * 30}\n
         You stupid! Haven't you ever heard of the Sphinx in a classic myth?
         Answering the Sphinx's questions wisely is the only way to live!!!
            \n        {'-' * 71}\n""")
            fail()     # Game over
        
        # Controlling unexpected inputs 
        else: 
            print(f"""\n{host_name}: You know that is not an option... Let's try again.\n""")
            sleep(0.7) # Establishing a pause

def game_3():
    words = ['Sphinx', 'Riddle', 'Day', 'Night']
    # Changing the order of words in the list
    random.shuffle(words)
    # Looping based on words in the list
    for word in words:
        # Breaking apart a word as characters
        characters = list(word)
        # Shuffling the characters
        random.shuffle(characters)
        # Setting shuffling characters as a word
        shuffled_word = ''.join(characters)
                
        guess = input("""\nWhat is the original form of {}: """.format(shuffled_word))
        
        if guess == word: 
            print("It is correct.")
            
        else: 
            print("It is incorrect.")
            
    # Showing the Sphinx's hint for the riddle      
    print("\n\nHint is {}".format(words))
    
    

################
## GAME OVER ##
###############

def fail():
    print(f"""\n\n{host_name}: Would you like to play again? (Yes/No)""")
    # Looping until a user input correctly
    while True:
        # Giving the choice between yes & no
        replay = input("> ")
        # Converting lower case
        replay = replay.lower()
        
        # Conditional statement to keep playing game or finish game
        if "yes" in replay:
            print(f"\n{host_name}: Great, let\'s go!\n")
            # Controlling the speed of user engagement
            input(prompt = "\n<Press ENTER to start again>\n")
            stage_1()  # Moving to stage 1
            
        elif "no" in replay:
            print(f"\n{host_name}: Thanks for playing {user_name}!")
            sys.exit() # The end
            
        # Controlling unexpected inputs     
        else:
            print(f"""\n{host_name}: I beg your pardon?""")
            sleep(0.7) # Establishing a pause
    

     
#####################
## Start the Game ##
####################

runGame()
```
